## Unificació de dataframes

```python
df = pd.merge(left=lloguer_df, right=soroll_df, on=["Codi_Districte", 'Nom_Districte', 'Codi_Barri', 'Nom_Barri'], how="outer")
df
```

## Depuració de dades

```python
#Eliminem els valors nuls per poder tenir un dataframe net
df.dropna(inplace=True)
df
```

```python
#Per poder dividir el dataframe farem una observació de quants valors n'hi ha de cada
df['Lloguer_mitja'].value_counts()
```

```python
#Farem ús de les máscares booleanes per fer la divisió del dataframe
lloguer_mensual = (df['Lloguer_mitja'] == 'Lloguer mitjà mensual (Euros/mes)')
lloguer_m2 = (df['Lloguer_mitja'] == 'Lloguer mitjà per superfície (Euros/m2 mes)')
```

```python
df_lloguer_mes = df[lloguer_mensual]
df_lloguer_mes
```

```python
df_lloguer_m2 = df[lloguer_m2]
df_lloguer_m2
```

```python
#Un cop tinguem els dos dataframes farem una agrupació per veure tant en un com en l'altre si hi ha alguna mena de relació entre els DB i els preu dels lloguers.
df_ag_lloguer_mes = df_lloguer_mes.groupby(['Nom_Districte', 'Nom_Barri', 'Rang_soroll'])[['Preu']].mean().sort_values(by='Rang_soroll', ascending=False)
df_ag_lloguer_mes
```

```python
df_ag_lloguer_m2 = df_lloguer_m2.groupby(['Nom_Districte', 'Nom_Barri', 'Rang_soroll'])[['Preu']].mean().sort_values(by='Rang_soroll', ascending=False)
df_ag_lloguer_m2
```
